```python
"""This code is used to randomly sample the CIFAR-10 dataset to obtain the corresponding three-class dataset."""
```

```python
num_class=3
```

```python
def dataset(inp,lab,num,NUM):
    L=len(lab)
    G_L=[0,1,8]
    r=np.zeros(num*num_class)
    for i in range(num_class):
        ma=np.zeros(NUM)
        ma[0:num]=1
        np.random.shuffle(ma)
        ma=np.where((ma==1))
        p_i=np.where((lab==G_L[i]))[0]
        p_i=np.array(p_i)
        p_i=p_i[ma]
        r[i*num:(i+1)*num]=p_i
    np.random.shuffle(r)
    r=r.astype('int64')
    x=inp[r]
    y=lab[r]
    for i in range(len(y)):
        if y[i]==8:
            y[i]=2
        if y[i]==9:
            y[i]=3
    return x,y
```

```python
def cifar10_dataset(y1_label=0,y2_label=1):
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
    x_train = x_train.astype('float32')
    x_test = x_test.astype('float32')
    x_train /= 255
    x_test /= 255
    x_tr,y_tr=dataset(x_train,y_train,1000,5000)
    x_te,y_te=dataset(x_test,y_test,1000,1000)
    return x_tr, y_tr, x_te, y_te
```

```python
x_train,y_train,x_test,y_test=cifar10_dataset()
```

```python
with open('data.pkl', 'wb') as f:
    pickle.dump([x_train,y_train,x_test,y_test],f)
```
