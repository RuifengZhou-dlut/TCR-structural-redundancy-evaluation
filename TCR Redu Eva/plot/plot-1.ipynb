```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
def load_progress(SAVE_FILE):
    with open(SAVE_FILE, "r") as f:
        return json.load(f)
```

```python
l='C:/Users/user/jupy/redu/'
```

```python
l1=['class-3/','class-3-dist/','class-10/']
```

```python
l2=["training_VGGNet16_log.json","training_ResNet18_log.json","training_ResNet32_log.json","training_ResNet56_log.json"]
```

```python
data=[]
for i in range(4):
    for j in range(3):
        SAVE_FILE=l+l1[j]+l2[i]
        data.append(load_progress(SAVE_FILE))
```

```python
plt.figure(1,figsize=(35,10))
x=np.arange(13)
plt.plot(x,data[0]['RR_L'][0],color='red',linewidth=3.0,marker='o',markersize='10',label='VGG-16')
plt.plot(x,data[1]['RR_L'][0],color='green',linewidth=3.0,marker='o',markersize='10',label='VGG-16+dist')
plt.plot(x,data[2]['RR_L'][0],color='blue',linewidth=3.0,marker='o',markersize='10',label='VGG-16/10')
plt.grid(True,linestyle='--',linewidth =2.0)
plt.xlabel('layers',size=50)
plt.xticks(range(13),labels=[str(i+1) for i in range(13)])
plt.ylabel('R$_{l}$',size=50)
plt.ylim(-0.45,1.05)
plt.tick_params(labelsize=35)
leg=plt.legend(bbox_to_anchor=(0.20,1.0),fontsize=35)
leg.get_frame().set_alpha(0)
plt.savefig('p-VGGNet16.pdf',dpi=200, bbox_inches='tight')
```
